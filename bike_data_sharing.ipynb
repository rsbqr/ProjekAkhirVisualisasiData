## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim apa rata rata terbanyak untuk pesepeda yang kasual dan musim apa yang rata ratanya sedikit pesepada yang kasual?

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    y="casual",
    x="season",
    data=hari_df.sort_values(by="casual",ascending=False),
)

plt.title("Bersepeda Kasual per Musim", loc="center", fontsize=15)
plt.ylabel('Rata Rata Bersepeda')
plt.xlabel('Musim')
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 2: Musim apa rata rata terbanyak untuk pesepeda yang terdaftar dan musim apa yang rata ratanya sedikit pesepada yang terdaftar?

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    y="registered",
    x="season",
    data=hari_df.sort_values(by="registered"),

)
plt.title("Bersepeda Terdaftar per Musim", loc="center", fontsize=15)
plt.ylabel("Rata Rata Bersepeda")
plt.xlabel("Musim")
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

## Conclusion

- Kesimpulan dari pertanyaan 1
Berdasarkan data rata rata dapat dilihat bahwa pada ***musim 1*** merupakan musim bersepeda kasual yang ***sedikit*** dan ***musim 3*** merupakan musim yang bersepeda kasual ***terbanyak***

- Kesimpulan dari pertanyaan 2
Berdasarkan data sebelumnya dapat dilihat bahwa pada musim yang sama pada ***musim 1*** merupakan musim bersepeda terdaftar yang ***sedikit*** dan ***musim 3*** merupakan musim yang bersepeda terdaftar ***terbanyak***


##RFM Analysis

```python
rfm_df_sum=hari_df.groupby(by="season").agg({
    "casual": ["sum"],
    "registered": ["sum"]
    })
rfm_df_sum.head()
```

```python
sum_casual_df = hari_df.groupby("season").casual.sum().sort_values(ascending=False).reset_index()
sum_casual_df.head()
```

```python
sum_registered_df = hari_df.groupby("season").registered.sum().sort_values(ascending=False).reset_index()
sum_registered_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="season", y="casual", data=sum_casual_df.head().sort_values(by="casual", ascending=True).head(), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Casual", loc="center", fontsize=15)
ax[0].tick_params(axis ='x', labelsize=12)


sns.barplot(x="season", y="registered", data=sum_registered_df.head().sort_values(by="registered", ascending=False).head(), palette=colors, ax=ax[1])
ax[1].set_ylabel("Jumlah Pesepeda")
ax[1].set_xlabel("Musim")
ax[1].set_title("Registered", loc="center", fontsize=15)
ax[1].tick_params(axis ='x', labelsize=12)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].tick_params(axis='x', labelsize=12)
'''
sns.barplot(x="quantity_x", y="product_name", data=rfm_df_sum.head().sort_values(by="quantity_x", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)'''

plt.suptitle("Perbandingan Jumlah Pesepeda Kasual dengan Pesepeda Terdaftar per Musim", fontsize=20)
plt.show()
```
